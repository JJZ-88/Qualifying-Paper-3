```python
## Run ChEES-HMC to tune parameters

def chees_tune(num_chains, target_log_prob, num_samples, init, key):
  #initialize HMC algo
  warmup = blackjax.chees_adaptation(target_log_prob, num_chains)
  key_warmup, key_sample = jax.random.split(jax.random.PRNGKey(key))
  optim = optax.adam(0.3) #set optimizer learning rate
  #.run actually runs the algorithm
  (last_states, parameters), _ = warmup.run(
      key_warmup,
      init, #initial positions
      0.1, #initial step size
      optim, #optimizer to use
      num_samples,   # num steps
  )
  #last states is the Wth iteration (last warmup step)
  #parameters are the optimal step size, trajectory, mass
  #create HMC kernel with tuned params
  kernel = blackjax.dynamic_hmc(target_log_prob, **parameters).step
  return kernel, last_states

```

```python
## Run HMC Sampler (in parallel) from tuned HMC kernel

def hmc_sampler(rng_key, kernel, initial_state, num_samples, num_chains):
    #allow for chain parallelization
    @jax.jit
    def one_step(states, rng_key):
        keys = jax.random.split(rng_key, num_chains)
        states, _ = jax.vmap(kernel)(keys, states)
        #vmap vectorizes, parallel chains but on 1 GPU. Pmap can go over multiple cores
        return states, states #lax.scan requires this output type

    keys = jax.random.split(rng_key, num_samples)
    #efficient for loop, returns carry (current) and accumulated values
    #running one_step based on number of iterations (provided by keys)
    all, states = jax.lax.scan(one_step, initial_state, keys)

    return all, states



```

```python
## Compute nested rhat from the (many) sampled chains at each time step n = 1,...,N

def compute_rhatv(chain):
  # print(chain)
  #chain is a 3d array. [n, m, k]
  dim = chain.shape
  superchain_means = chain.mean(-1) #mean of each superchain for each time step
  superchain_means_n = (jnp.cumsum(superchain_means, 0) / jnp.arange(1, superchain_means.shape[0] + 1)[:, jnp.newaxis])
  #mean of each superchain from time 1:n, n (newaxis just gives dummy axis to match dimensions)
  # print(superchain_means)
  # print(superchain_means_n)
  # print(superchain_means.shape)
  # print(np.mean(chain[0][0]))
  # print(np.mean(chain[0][31]))
  # print(np.var(superchain_means[0]))

  # print(superchain_means)
  # print(superchain_means_n)
  # print(np.mean(chain[0,1,:]))
  Bv = superchain_means_n.var(1) #between superchain variance for each n

  subchain_means_n = (jnp.cumsum(chain, 0) / jnp.arange(1, chain.shape[0] + 1)[:, jnp.newaxis, jnp.newaxis])
  #mean of each subchain from time 1:n, n = 1,...,n
  Bk = subchain_means_n.var(2) if dim[2] > 1 else jnp.zeros((dim[0], dim[1]))
  #between subchain variance (set to 0 if M = 1) for each n

  sum_sq_n = jnp.cumsum(jnp.square(chain), 0)
  subchain_mean_sq_n = jnp.square(subchain_means_n) * jnp.arange(1, chain.shape[0] + 1)[:, jnp.newaxis, jnp.newaxis]
  Wk_full2 = (sum_sq_n - subchain_mean_sq_n) / jnp.arange(0, chain.shape[0])[:, jnp.newaxis, jnp.newaxis]
  #get within subchain variance var(f^nmk) = (E((f^nmk)^2) - E(f^nmk)^2)/(N - 1)

  Wv = (jnp.concatenate([jnp.zeros((1, dim[1])), Wk_full2.mean(2)[1:(dim[0]), :]]) + Bk).mean(1)
  #concatenate a 0 column (for N = 1) and get within superchain variance

  Rhatv = jnp.sqrt(1 + Bv / Wv)
  #compute nest rhat and return each component

  superchain_var = chain.var(1)
  pers_var = superchain_var.mean(1)
  return Bv, Wv, Rhatv, pers_var
```

```python
## Problem setup
n_chains = 2048
n_super_chains = 32
n_samples = 1
n_sub_chains = n_chains // n_super_chains #floor division
```

```python
## Run chains for different warmup lengths and return non-stationary variance, bias

def run_chains(warm_arr, init_fn, target_fn, ndim, n_chains, n_sub_chains, n_super_chains, n_samples, true_val):
  nw = warm_arr.shape[0]
  bcv = np.zeros(nw)
  bias = np.zeros(nw)
  pv = np.zeros(nw)
  rhat = np.zeros(nw)

  for i in range(nw):
    print(warm_arr[i])
    init = jnp.repeat(init_fn(jax.random.PRNGKey(warm_arr[i]), (n_super_chains, ndim)), n_sub_chains, axis = 0)
    kernel, last_states = chees_tune(n_chains, target_fn, warm_arr[i], init, 3)
    hmc = hmc_sampler(jax.random.PRNGKey(warm_arr[i]), kernel, last_states, n_samples, n_chains)
    b, w, r, pers_var = compute_rhatv(hmc.position[:,:,0].reshape(n_samples, n_super_chains, n_sub_chains))
    pv[i] = w[0]
    bias[i] = (hmc.position[:,:,0].mean() - true_val) ** 2
    bcv[i] = b[0]
    rhat[i] = r[0]
    del hmc, kernel, last_states, init

  return bcv, bias, pv, rhat


```

```python
def run_chains2(init_fn, target_fn, ndim, n_chains, n_sub_chains, n_super_chains, true_val, warmup, iters):
  bcv = np.zeros(iters)
  bias = np.zeros(iters)
  pv = np.zeros(iters)
  rhat = np.zeros(iters)

  init = jnp.repeat(init_fn(jax.random.PRNGKey(88), (n_super_chains, ndim)), n_sub_chains, axis = 0)
  kernel, last_states = chees_tune(n_chains, target_fn, warmup, init, 8)
  all, state = hmc_sampler(jax.random.PRNGKey(5005), kernel, last_states, 1, n_chains)
  for i in range(iters):
    print(i)
    all, state = hmc_sampler(jax.random.PRNGKey(i), kernel, all, 1, n_chains)
    # print(all)
    # print(state.position.shape)
    # print(state.position[:,:,0].reshape(1, n_super_chains, n_sub_chains))
    # print(state.position[:,:,0].reshape(1, n_super_chains, n_sub_chains).shape)
    # print(state.position[:,:,0].reshape(1, n_super_chains, n_sub_chains)[0][0].shape)
    b, w, r, pers_var = compute_rhatv(state.position[:,:,0].reshape(1, n_super_chains, n_sub_chains))
    pv[i] = w[0]
    bias[i] = (state.position[:,:,0].mean() - true_val) ** 2
    bcv[i] = b[0]
    rhat[i] = r[0]

  del kernel, last_states, init, all, state
  output.clear()
  return bcv, bias, pv, rhat

```
